*Answer to question 1*

###Distribution-based clustering:

•	Gaussian Mixture Models (GMMs): GMMs model the data as a mixture of Gaussian distributions. Each cluster is represented by a Gaussian distribution, and the algorithm assigns each data point to the cluster with the highest probability.

###Model-based clustering:

•	Spectral Clustering: Spectral Clustering is a clustering algorithm that uses the spectral properties of the data to find clusters. The algorithm constructs a similarity matrix between the data points, and then uses the eigenvectors of the similarity matrix to find clusters.

###How these algorithms work:

GMMs work by first estimating the parameters of the Gaussian distributions that make up the mixture model. This is done using the Expectation-Maximization algorithm. Once the parameters of the Gaussian distributions have been estimated, each data point is assigned to the cluster with the highest probability.

Spectral Clustering works by first constructing a similarity matrix between the data points. The similarity matrix represents the similarity between each pair of data points. Then, the algorithm uses the eigenvectors of the similarity matrix to find clusters. The algorithm clusters the data points by assigning them to the clusters with the same principal components.

In summary, these are just a few of the many clustering algorithms that are available. The best algorithm to use will depend on the specific data set and the desired results.

## KMeans Clustering

```python
''' Elbow Method to identify the Number of Clusters
using the --Within-CLuster Sum of Squared--(WSS)'''

import sklearn.cluster as cluster

k_range = range(1,11) #Range of k=10 clusters values to choose from
wss = []              # Initialize WSS as an empty class to store the values

for k in k_range:
  kmean = cluster.KMeans(n_clusters = k, init= 'k-means++', max_iter=50, n_init=10)
  kmean_fit = kmean.fit(Xtrain_scal)
  wss_iter = kmean.inertia_
  wss.append(wss_iter)     # Store the WSS scores
  print(k)
```

```python
# Store the Number of Clusters along with WSS Scores in a DataFrame
mycenter = pd.DataFrame({'Cluster':k_range, 'Wss':wss})
print(mycenter)
```

```python
# Plotting the WSS vs Number of clusters
import matplotlib.pyplot as plt

plt.plot(k_range, wss, 'r-*')
plt.xlabel('Number of cluster K')
plt.ylabel("WSS scores")
plt.title('Elbow Method')
plt.show()
print('\nThe Elbow Method shows K = 2 ')
```

```python
''''Silhouette Method to identify the Number of Clusters,
    using Silhouette Scores'''

from sklearn import metrics
from sklearn.metrics.cluster import silhouette_score

#Silhouette
j_range= range(2,11)    #Range of j=10 clusters values to choose from
silhouette_scores = []  #Initialize Scores as an empty class to store the values


for j in j_range:
  label = cluster.KMeans(n_clusters = j, init='k-means++', random_state=10, n_init='auto')
  cluster_label = label.fit_predict(Xtrain_scal)
  silhouett_avg = silhouette_score(Xtrain_scal, cluster_label, sample_size=100,metric='euclidean')

  silhouette_scores.append(silhouett_avg)  # Store the silhouette scores
  print('For Cluster=',
        j, 'The avg Silhoute Score is:',
        silhouett_avg)
```

```python
#Plot the Silhouette Scores
plt.plot(j_range, silhouette_scores, 'b--')
plt.xlabel('Number of cluster (K)')
plt.ylabel('Silhouette SCore')
plt.title('Silhouette Method')
plt.show()
```

```python
# Choose the number of clusters with the highest silhouette score
k_opt = j_range[np.argmax(silhouette_scores)]  # Select the Silhouette score

print('The highest average Silhouette Score = 0.22312')
print('Therefore, the Number of Clusters are:', k_opt)
```

 ## K Means Clustering with 2 clusters chosen from Elbow and Silhouette Method

```python
"""### K Means Clustering Algorithm with 2 clusters"""
#KMeans Method
k_means = cluster.KMeans(n_clusters=2, init ="k-means++", random_state=0, n_init="auto")
k_means.fit(Xtrain_scal)

ypred = k_means.labels_  # x labels = y: Number 0f Clusters,
print(ypred)
ypred.shape
```

##**Note:** I wanted to see what would be the performance since we had y train data given

```python
# # Evaluate the performance of the K Means Clustering model using the Recall_score
from sklearn.metrics import recall_score

xtrn_kmeans_rcall = recall_score(ypred, y_train, average='weighted')
print('The mean Recall accuracy of the model predictions is:', round(100*xtrn_kmeans_rcall,2),'%')

# Evaluate the model performance of the model using the Score metric
from sklearn.metrics import f1_score

xtrn_kmeans_scor1 = f1_score(ypred, y_train)
print('\nThe mean f1 accuracy of the model predictions is:', round(100*xtrn_kmeans_scor1,2),'%')
```

##Agglomerative (Hierarchical) Clustering

```python
'''Let us draw a Dendrogram to look for the number of Clusters'''
#Import the necessary libraries
import scipy.cluster.hierarchy as sch

plt.figure(figsize = (10, 6))  ##Plotting
dendogrm = sch.dendrogram(sch.linkage(Xtrain_scal, method='ward'))
plt.show()
```

##Agglomerative (Hierarchical) Clustering with 2 clusters

```python
"""Agglomerative Clustering Algorithm with 2 clusters"""

from sklearn.cluster import AgglomerativeClustering
import numpy as np

# Agglomerative Clustering Method
aglom_clus = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
aglom_clus.fit(Xtrain_scal)
y_agm_pred = aglom_clus.labels_
print(y_agm_pred)
```

##Performance of the Model

```python
# Evaluate the performance of the Agglomerative Clustering model using the Score metric
xtrn_aglom_scor1 = f1_score(y_agm_pred, y_train)
print('The mean f1 accuracy of the Agglomerative Clustering model is:', round(100*xtrn_aglom_scor1,2),'%\n')

# Recall_score Agglomerative Clustering
xtrn_aglom_rcall = recall_score(y_agm_pred, y_train, average='weighted')
print('The mean Recall accuracy of the Agglomerative Clustering model is:', round(100*xtrn_aglom_rcall,2),'%')
```
